```python
import pandas as pd
data_path = "/home/jimmy/LaTeX/markov_project/code/data/Overall_us_2000-2017.csv"
```

```python
df = pd.read_csv(data_path)
actual = df.columns[-3]
to2011 = df.columns[-2]
to2017 = df.columns[-1]
df
```

```python
f11to17 = (df.loc[6:9][to2017] / df.loc[6:9][to2011]).mean()
f17to11 = (df.loc[6:9][to2011] / df.loc[6:9][to2017]).mean()

df.loc[:5, to2017] = (df.loc[:5, to2011] * f11to17).round(1)
df.loc[10:, to2011] = (df.loc[10:, to2017] * f17to11).round(1)

df
```

```python
fires = "Fires (in billions)"
loss_prefires = "Loss (in billions) (pre-fires) to2017"
df[fires] = df["Fires (in thousands)"] / 1000
df[loss_prefires] = df[to2017] / df[fires]
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# 设置参数
lambda_values = df[loss_prefires]  # 不同泊松分布的参数
years = df["Year"]
# 生成 x 值
x = np.arange(0, 100, 0.1)
plt.figure(figsize=(20, 10))

# 绘制每个泊松分布的概率质量函数
for i, lam in enumerate(lambda_values):
    # 计算概率质量函数
    y = poisson.pmf(x, lam)

    # 绘制泊松分布
    plt.plot(x, y, label='λ = {}, years = {}'.format(round(lam,1), years[i]))

# 添加图例、标签和标题
plt.legend()
plt.xlabel('X')
plt.ylabel('Probability')
plt.title('Poisson Distribution')

# 显示图形
plt.show()
```
